```python
import matplotlib.pyplot as ptl

fig, axes = ptl.subplots(1,3, figsize=(14, 4))

sns.scatterplot(data = df, x = "TV", y="sales", ax = axes[0])
sns.scatterplot(data = df, x = "radio", y="sales", ax = axes[1])
sns.scatterplot(data = df, x = "newspaper", y="sales", ax = axes[2])

axes[0].set(title = "TV vs sales")
axes[1].set(title = "radio vs sales")
axes[2].set(title = "newspaper vs sales")
```

zip tp loop throught both of them as the same time

```python
fig, axes = ptl.subplots(1,3, figsize=(14,4))

features = ("TV", "radio", "newspaper")
for freature, ax in zip(features, axes):
    sns.scatterplot(data=df, x=freature, y="sales", ax=ax)
```
