```python
class Person:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age 
    def __repr__(self):
        return f"Person(name={self.name}, gender={self.gender}, age={self.age})"


p1 = Person("bl채", "m", 34)
p1
```

```python
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    gender: str
    age : int

p5 = Person(name = "bbbb", gender = "f", age = 3)
p5
```
