#Linear regression 

```python
import pandas as pd

df = pd.read_csv("Advertising.csv", index_col=0)
df
```

```python
df.shape # amount of colums and rows
```

```python
x, y = df.drop("sales", axis = 1), df["sales"]
x.head() # all the  features
```

```python
y.head() #labes
```

#Test|test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)

x_train
```

```python
x_test
x_test.shape
```

```python
y_train.shape, y_test.shape
```

## Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_train)

scaler
```

```python
scaled_x_train = scaler.transform(x_train)
scaled_x_test = scaler.transform(x_test)

scaled_x_train, scaled_x_train.shape 
```

```python
scaled_x_train.min(), scaled_x_train.max()
```

```python
scaled_x_test.min(), scaled_x_test.max()
```

## Linear regression 

```python
from sklearn.linear_model import LinearRegression 

model = LinearRegression()
model 
```

```python
model.fit(scaled_x_train, y_train)
model.coef_
```

```python
model.intercept_
```

## Prediction

```python
test_sample_features = scaled_x_test[0].reshape(1,-1)
test_sample_target = y_test.values[0]
test_sample_features, test_sample_target
```

```python
model.predict(test_sample_features)
```

```python
test_sample_target
```

### prediction on test data

```python
y_pred = model.predict(scaled_x_test)
y_pred.shape
```
