b) Draw a line chart of quantity for each year. Is there a year that should be skipped?

```python
df_b = (df_0.groupby("Year")["Quantity"].sum().reset_index())
```

```python
plt.figure(figsize=(15,12))
plt.plot(df_b["Year"], df_b["Quantity"])

plt.xlabel("Year")
plt.ylabel("Quantity")
plt.title("Quantity per year")

plt.grid(True)
plt.show
plt.savefig("quantait_per_year.png")
```

c) Draw a line chart of average CO2 emissions for same years that as in b)

```python
df_c = (df_0.groupby("Year", as_index=False)["Avg_CO2"].mean())

plt.figure(figsize=(10, 6))
plt.plot(df_c["Year"], df_c["Avg_CO2"])

plt.xlabel("Year")
plt.ylabel("Avg_CO2")
plt.title("CO2 per year")

plt.grid(True)
plt.show()
```
